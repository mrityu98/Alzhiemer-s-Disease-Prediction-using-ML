```python
import numpy
import math
count=df_main.shape[0]
abeta_val1,abeta_val2,abeta_val3,label1,label2,label3=[],[],[],[],[],[]
for i in range(count):
    if type(df_main['DX'][i]) == str and (type(df_main['ABETA'][i]) == float and ~math.isnan(df_main['ABETA'][i])):
    
        if df_main['DX'][i]=='CN':
            abeta_val1.append(df_main['ABETA'][i])
            label1.append(0)
        elif df_main['DX'][i]=='Dementia':
            abeta_val2.append(df_main['ABETA'][i])
            label2.append(1)
        elif df_main['DX'][i]=='MCI':
            abeta_val3.append(df_main['ABETA'][i])
            label3.append(2)
```
