```python
# Data for plotting
import matplotlib.pyplot as plt
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, (y[i]+0.9000)/2, y[i], ha = 'center',color='white')
models = [ 'XGBoost','Random Forest','GBoost','Log_Regression','SVM','K-NN']
accuracies = [0.9592,0.9552,0.9527,0.9481,0.9479,0.9388]  # Retrieved from your results
#accuracies.sort(reverse=True)
# Plotting the bar graph
plt.figure(figsize=(10, 5))
plt.bar(models, accuracies, color=['green'])
addlabels(models, accuracies)
plt.ylim(0.90, 0.98)
plt.ylabel('Accuracy')
plt.title('Average accuracy comparison of different classifiers with a 70:30 train_test split for 5 runs')
plt.savefig('accuracies_final.png')
plt.show()
```
